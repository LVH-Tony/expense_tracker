```python
df = pd.read_csv('Books Distribution Expenses - Books.csv')

# CLEANING
df.replace('python', 'Python', inplace=True)
df.loc[df.title == 'Python for R Users', 'category'] = 'Python'
df.loc[df.title == "Joe Celko's SQL for Smarties", 'category'] = 'SQL'
df.loc[df.title == 'Essentials of Data Science and Analytics', 'category'] = 'data science'
df.loc[df.title == "Machine Learning for Algorithmic Trading", 'category'] = 'data science'
df.loc[df.title == 'Hands-On Data Science for Marketing', 'category'] = 'data science'
df.loc[df.title == "Data Science at the Command Line", 'category'] = 'data science'
df.loc[df.title == "Data Science for Business", 'category'] = 'data science'
df.loc[df.title == 'Marketing Data Science', 'category'] = 'data science'
df.loc[df.title == 'Taming Text', 'category'] = 'NLP'
df.loc[df.title == "Learning RStudio for R Statistical Computing", 'category'] = 'R Studio'
df.loc[df.title == "Functional Programming in R", 'category'] = 'R Studio'
df.loc[df.title == "Software for Data Analysis", 'category'] = 'R Studio'
df.loc[df.title == "bookdown", 'category'] = 'R Studio'
df.loc[df.title == "Statistics for Data Science", 'category'] = 'statistics'
df.loc[df.title == "S Programming", 'category'] = 'statistics'

df.drop_duplicates(inplace=True)

# published_date -> pd datetime format m/d/y error coerce means if error, return NaT
df['published_date'] = pd.to_datetime(df['published_date'], dayfirst=True, errors='coerce', format='%m/%d/%Y')

df.loc[df.id == 'UOM:39015056926119', 'published_date'] = '1879-01-01'

print(df.shape)
df.sample(10)
```
